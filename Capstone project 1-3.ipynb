# Question 3 <br>

**Filtering rows that include "Toronto" in their "Borough" column**

```python
result_Toronto=result[result['Borough'].str.contains("Toronto")]
result_Toronto.head()
```

**Using dummy variable to change "Borough" column**

```python
result_Toronto_onehot=pd.get_dummies(result_Toronto[['Borough']])
```

```python
result_Toronto_onehot['Latitude']=result_Toronto['Latitude']
result_Toronto_onehot['Longitude']=result_Toronto['Longitude']
result_Toronto_onehot['Neighborhood']=result_Toronto['Neighborhood']

# move neighborhood column to the first column
fixed_columns = [result_Toronto_onehot.columns[-1]] + list(result_Toronto_onehot.columns[:-1])
result_Toronto_onehot = result_Toronto_onehot[fixed_columns]
result_Toronto_onehot.reset_index(inplace=True,drop=True)
```

**Importing required libraries**

```python
# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# Numpy
import numpy as np

# import k-means from clustering stage
from sklearn.cluster import KMeans

!pip -q install folium
import folium # map rendering library

!pip -q install geopy 
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values


print('Libraries imported.')
```

```python
# set number of clusters
kclusters = 5

Toronto_grouped_clustering = result_Toronto_onehot.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(Toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels
result_Toronto_onehot.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
result_Toronto_onehot.head()
```

**Showing clusters on the map**

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Manhattan are {}, {}.'.format(latitude, longitude))
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(result_Toronto_onehot['Latitude'], result_Toronto_onehot['Longitude'], result_Toronto_onehot['Neighborhood'], result_Toronto_onehot['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
